# Drop Unwanted Columns

# Remove Rating score and start from Society column  

```python
df["society"].value_counts()
```

```python
import re

df['society'] = (
    df['society']
    .astype(str)
    .str.replace(r'\d+(\.\d+)?\s?★', '', regex=True)
    .str.strip()
    .str.lower()
)
```

```python
df["society"].value_counts()
```

```python
df["society"].value_counts().shape
```

```python
df["society"] = df["society"].str.replace("nan","independent")
```

# convert Price In CR

```python
df = df[df["price"] != "price on Request"]
```

```python
df.sample(10)
```

```python
df["price"].value_counts()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

# price_per_square_fit

```python
df['price_per_squft'].value_counts()
```

```python
df['price_per_squft'] = (
    df['price_per_squft']
    .str[0]                        # Access first element of the list
    .str.replace(r'[^\d]', '', regex=True) # Remove all non-numeric characters

)
```

```python
df["price_per_squft"].sample()
```

```python
df["areaWithType"].value_counts()
```

```python
df["bedRoom"].value_counts()
```

# remove Text From Columns

```python
df = df[~df["bedRoom"].isnull()]
```

```python
df["bedRoom"] = df["bedRoom"].str.split(" ").str.get(0).astype("int")
```

```python
df["bathroom"] = df["bathroom"].str.split(" ").str.get(0).astype("int")
```

```python
df["balcony"] = df["balcony"].str.split(" ").str.get(0).str.replace("No","0")
```

# Additinal Room

```python
df["additionalRoom"].fillna("not avilable",inplace=True)
```

```python
df["additionalRoom"] = df["additionalRoom"].str.lower()
```

# Floor

```python
df["noOfFloor"].sample(10)
```

```python
df['noOfFloor'] = (
    df['noOfFloor']
    .str.split(' ')
    .str.get(0)
    .replace({'Ground': '0', 'Basement': '-1', 'Lower': '0'})
    .str.extract(r'(-?\d+)') # Added -? to capture the negative sign for Basement
    .astype(float)
)
```
